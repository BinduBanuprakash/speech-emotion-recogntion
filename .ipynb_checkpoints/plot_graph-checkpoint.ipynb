```python
import librosa
import librosa.display
import sklearn
import matplotlib.pyplot as plt 

class plot_wave_graph():
    def __init__(self, filename):
        self.data, self.sampling_rate = librosa.load(filename)
        self.plot_waveform()
        self.spectogram()
#         self.normalised_waveform()
        
    def plot_waveform(self):
        """display waveform"""
        fig = plt.figure(figsize=(10, 5))
        plt.title("WAVEFORM")
        librosa.display.waveplot(self.data, sr=self.sampling_rate)
        plt.show()
        fig.savefig('wave.png')
        
    def spectogram(self):
        """display Spectrogram"""
        X = librosa.stft(self.data)
        Xdb = librosa.amplitude_to_db(abs(X))
        fig = plt.figure(figsize=(10, 5))
        plt.title("WAVE SPECTOGRAM")
        librosa.display.specshow(Xdb, sr=self.sampling_rate, x_axis='time', y_axis='hz') 
        plt.colorbar()
        plt.show()
        fig.savefig('wave_spectogram.png')
        
#     def normalize(self,x, axis=0):
#         """Normalising the spectral centroid for visualisation"""
#         return sklearn.preprocessing.minmax_scale(x, axis=axis)
        
#     def normalised_waveform(self):
        
#         spectral_centroids = librosa.feature.spectral_centroid(self.data, sr=self.sampling_rate)[0]
#         spectral_centroids.shape
        
#         fig = plt.figure(figsize=(10, 5))
#         plt.title("NORMALISED VERSION")
        
#         # Computing the time variable for visualization
#         frames = range(len(spectral_centroids))
#         t = librosa.frames_to_time(frames)
        
#         #Plotting the Spectral Centroid along the waveform
#         librosa.display.waveplot(self.data, sr=self.sampling_rate, alpha=0.4)
#         plt.plot(t, self.normalize(spectral_centroids), color='r')
#         spectral_rolloff = librosa.feature.spectral_rolloff(self.data, sr=self.sampling_rate)[0]
#         librosa.display.waveplot(self.data, sr=self.sampling_rate, alpha=0.4)
#         plt.show()
#         fig.savefig('normalised_version.png')
    
    def mfcc(self,mel_coefficients):
        S = librosa.feature.melspectrogram(self.data, sr=self.sampling_rate, n_mels=128)
        # Convert to log scale (dB). We'll use the peak power as reference.
        log_S = librosa.amplitude_to_db(S, ref=np.max)
        # Make a new figure
        fig = plt.figure()
        # Display the spectrogram on a mel scale
        # sample rate and hop length parameters are used to render the time axis
        librosa.display.specshow(log_S, sr=self.sampling_rate, x_axis='time', y_axis='mel')
        # Put a descriptive title on the plot
        plt.title('mel power spectrogram')
        # draw a color bar
        plt.colorbar(format='%+02.0f dB')
        # Make the figure layout compact
        plt.tight_layout()
        fig.savefig('mel_spectogram.png')
```
